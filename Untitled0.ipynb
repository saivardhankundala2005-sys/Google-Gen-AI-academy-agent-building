```python
import os
import pandas as pd
import kagglehub
path=kagglehub.dataset_download("crawford/emnist")
train_data=pd.read_csv(os.path.join(path,"emnist-letters-train.csv"))
test_data=pd.read_csv(os.path.join(path,"emnist-letters-test.csv"))
```

```python
x_train = train_data.values[: , 1:]
y_train =train_data.values[:,0]-1

x_test = test_data.values[:,1:]
y_test = test_data.values[:,0]-1
```

```python
import numpy as np

x_train =x_train.reshape(-1,28,28,1)
x_test =x_test.reshape(-1,28,28,1)

x_train =np.rot90(x_train, k=-1,axes=(1,2))
x_train =np.flip(x_train,axis =2)

x_test =np.rot90(x_test,k=-1,axes =(1,2))
x_test =np.flip(x_test,axis = 2)
```

```python
import tensorflow as tf

x_train =x_train.astype("float32")/255
x_test =x_test.astype("float32")/255

y_train =tf.keras.utils.to_categorical(y_train,26)
y_test =tf.keras.utils.to_categorical(y_test,26)

print(y_test)
```

```python
from tensorflow.keras import models,layers

model = models.Sequential([
    layers.Conv2D(32,(3,3),activation='relu',input_shape=(28,28,1)),
    layers.MaxPooling2D(2,2),

    layers.Conv2D(64,(3,3), activation='relu'),
    layers.MaxPooling2D(2,2),

    layers.Flatten(),
    layers.Dense(128,activation ='relu'),
    layers.Dense(26,activation= 'softmax')

])
```

```python
model.compile(optimizer='adam',loss ='categorical_crossentropy' , metrics=['accuracy'])
```

```python
model.fit(x_train,y_train , epochs =5,batch_size =128,validation_split=0.2)
```

```python
loss,accuracy =model.evaluate(x_test, y_test)
print(f"loss ={loss}")
print(f"accuracy ={accuracy}")
```
